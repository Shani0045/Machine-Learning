```python
from numpy import array,mean
```

```python
x=array([1,2,3,4,5])
y=array([1,3,5,7,9])
```

```python
def lin(dx,dy):
    m=((mean(dx)*mean(dy))-mean(dx*dy))/((mean(dx)**2)-mean(dx**2))
    c=mean(dy)-m*mean(dx)
    return m,c
lin(x,y)
```

```python
m,c=lin(x,y)
```

```python
ry=[m*i+c for i in x]
ry
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
plt.scatter(x,y,marker='X',s=150)
plt.plot(x,ry,"--")
plt.plot(x,[mean(y) for _ in x],":")
plt.show()
```
