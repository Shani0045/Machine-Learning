```python
import matplotlib.pyplot as plt
plt.style.use('_classic_test')
# plt.style.available    # it is use for check the all style
```

```python
import numpy as np
```

```python
x=np.arange(1,50,0.1)
y=np.sin(x)
plt.plot(x,y,':',label='sin',color='red')
plt.plot(x,np.cos(x),'-.',label='cos',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sin graph')
plt.legend(loc='upper right')  # loc means location of your legend
plt.show()
```

```python
x=[1,4,3,2,7,6]
y=[5,9,4,10,3,4]
plt.scatter(x,y,marker='*',s=500,alpha=1,edgecolors='yellow')  # s means size 
plt.show()
```

```python
p=[1,2,4,3,5,6,12,23,17,18,12,13,14,4,12,12,13,24,25,26,27,28,29,1,2,1,2,3,4,67]
b=[1,5,10,15,20,25,30]
plt.hist(p,b,histtype='barstacked',rwidth=0.5)   
plt.show()
```

```python
x=['australia','india','newzeland','england','pakistan']
pt=[14,11,11,10,9]
x1=[0,1,2,3,4]
plt.bar(x1,pt,tick_label=x)  #tick_label means name show karega niche x ka
plt.show()
```

```python
plt.pie(pt,labels=x,explode=(.1,0,0,0.2,0),autopct='%0.2f',colors=['b','g','y','r','c'],shadow=True)
plt.show()  # explode anticlockwise chalta hai jiske value high karenge vo bahar nikal jayega
# autopct percentage ko show karega 
```

```python
x=[1,2,3,4,5]
y1=[5,3,6,5,8]
y2=[4,7,9,8,3]
y3=[7,5,6,8,9]
plt.stackplot(x,y1,y2,y3,colors=['y','g','c'])
plt.show()
```

```python
p=plt.subplot(2,2,1)
p.plot(x,y1)
p1=plt.subplot(2,2,2)
p1.scatter(x,y1)
p2=plt.subplot(223)
p2.bar(x,y1)
p3=plt.subplot(224)
p3.pie(y1,)
plt.show()

#where  2,2,1 
#2 is row , 2 is column and 1 is no of box where plot is made
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
p=plt.subplot(221)
plt.title("bar graph")
p.bar(x,y)
q=plt.subplot(2,2,2)
plt.title("plot graph")
q.plot(x,y)
r=plt.subplot(2,2,3)
plt.title("pie graph")
r.pie(y1)
s=plt.subplot(2,2,4)
plt.title('scatter graph')
s.scatter(x,y)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib notebook
a=plt.figure()
p=a.add_subplot(111,projection='3d')
p.scatter(x,y1)
plt.title('3-D graph plot')
p.set_xlabel('x-axis')
p.set_ylabel('y-axis')
p.set_zlabel('z-axis')
plt.show()
```
