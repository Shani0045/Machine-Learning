```python
from bs4 import BeautifulSoup
```

```python
from requests import get
```

```python
url="https://www.flipkart.com/"
```

```python
data=get(url)
```

```python
print(data)
```

```python
soup=BeautifulSoup(data.text,"html.parser")
```

```python
print(soup.prettify())
```

```python
import re
```

```python
for data in soup.find_all("img",{'src':re.compile('.jpg')}):
    print(data["src"]+"\n")
```
