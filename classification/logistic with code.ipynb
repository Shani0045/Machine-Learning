```python
import numpy.matlib as ml
def cost(x,y,th):
    t=ml.matrix(th)
    xm=ml.matrix(x)
    s=sigmoid(t*xm.T)
    first=y*np.array(np.log(s))
    second=(1-y)*np.array(np.log(1-s))
    return -1*(first+second).mean()
cost(x_tr,y_tr,th)
```

```python
def gradient(x,y,th):
    t=ml.matrix(th)
    xm=ml.matrix(x)
    s=sigmoid(th*xm.T)
#     print(x.shape)
    err=s-y
    return err.shape
    
```

```python
def gradient(x,y,th):
    t=ml.matrix(th)
    xm=ml.matrix(x)
    s=sigmoid(th*xm.T)
#     print(x.shape)
    err=s-y
    n=np.array(x)*np.array(err).T
    th=th-0.001*n.mean(axis=0)
    return th
    
```
