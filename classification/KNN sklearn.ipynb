```python
data=pd.read_csv('ucicancer.csv')
data.head()
```

```python
data.columns
```

```python
data.columns=['a','b']
```

```python
data.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
data.columns
```

```python
import numpy as np
data.dtypes
data.g=data.g.replace('?',15)
data.g=data.g.astype(dtype=np.int64)
```

```python
data.dtypes
```

```python
data.drop('a',axis=1,inplace=True)
x=data.drop('class',axis=1)
y=data['class']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.2,random_state=0)
x_tr.shape
```

```python
x_ts
```

```python
algo=KNeighborsClassifier(n_neighbors=5,metric='euclidean')#n_neighbors=5,n_jobs=2)
algo.fit(x_tr,y_tr)
score=algo.score(x_ts,y_ts),
pr=algo.predict(x_ts)
```
