```python
data=pd.read_csv('data.csv')
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
data.columns
```

```python
x=data.iloc[:,0:48]
y=data.iloc[:,-1]
x
```

```python
from sklearn.model_selection import train_test_split
# from sklearn.metrics import accuracy_score
```

```python
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.33,random_state=17)
```

```python
bernNB=BernoulliNB()
```

```python
bernNB.fit(x_tr,y_tr)
```

```python
bernNB.score(x_ts,y_ts)
```

```python
# bernNB.r2_score(x_ts,y_ts)
```

```python
multinb=MultinomialNB()
multinb.fit(x_tr,y_tr)
multinb.score(x_ts,y_ts)
```

```python
multinb.predict(x_ts)
```

```python
gaussnb=GaussianNB()
gaussnb.fit(x_tr,y_tr)
print(gaussnb.score(x_ts,y_ts))
gaussnb.predict(x_ts)
```
