```python
def fun(prd,d):
    cl=class_prob.copy()
    for i in range(len(prd)):
        for j in d:
            cl[j]+=d[j][i].get(prd.iloc[i],0)
    return max(cl,key=cl.get)
fun(x_ts.iloc[0],d)
```

```python
c=0
for i in range(x_ts.shape[0]):
    if fun(x_ts.iloc[i],d)==y_ts.iloc[i]:
        c+=1
print(c)
c/x_ts.shape[0]
```
