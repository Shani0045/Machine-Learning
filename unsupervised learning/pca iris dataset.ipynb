```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.keys()
```

```python
print(iris.DESCR)
```

```python
x=iris.data
y=iris.target
```

```python
dt=pd.DataFrame(x,columns=iris.feature_names)
```

```python
dt.head()
```

```python
dt['target']=y
```

```python
dt
```

```python
dt['flower']=dt.target.apply(lambda x:iris.target_names[x])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_scaled=sc.fit_transform(x)
```

```python
x_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.95)
```

```python
principal_components=pca.fit_transform(x_scaled)
```

```python
df=pd.DataFrame(data=principal_components,columns=['principal componets 1','principal comonent 2'])
```

```python
df
```

```python
final_df=pd.concat([df,dt[['target']]],axis=1)
```

```python
final_df
```

```python
pca.explained_variance_ratio_
```

```python
final_df.iloc[:,0]
```

```python
plt.figure(figsize=(8,8))
plt.scatter(final_df.iloc[:,0],final_df.iloc[:,1],c=iris.target,cmap='viridis',s=50)
plt.style.use('ggplot')
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.title('2 components of PCA')
```
