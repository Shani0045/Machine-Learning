```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=load_iris()
```

```python
data.target
```

```python
dt=pd.DataFrame(data.data,columns=('sepal length','sepal width','petal length','petal width'))
dt
```

```python
dt
```

```python
dt.drop(['sepal length','sepal width'],axis=1,inplace=True)
```

```python
dt.head()
```

```python
kmean=KMeans(n_clusters=3)
```

```python
cluster=kmean.fit_predict(dt)
```

```python
dt['cluster']=cluster
```

```python
dt
```

```python
dt1=dt[dt['cluster']==0]
dt2=dt[dt['cluster']==1]
dt3=dt[dt['cluster']==2]
```

```python
%matplotlib inline
plt.figure()
```

```python
plt.style.use('classic')
plt.scatter(dt1['petal length'],dt1['petal width'],c='y')
plt.scatter(dt2['petal length'],dt2['petal width'])
plt.scatter(dt3['petal length'],dt3['petal width'])
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.title('iris dataset kmeans')
```

```python
plt.savefig('iris kmean.jpg')
```
