# KMeans

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dt=pd.read_csv('Mall_Customers.csv')
```

```python
dt.columns
dt.head()
```

```python
x=dt.iloc[:,[3,4]]
x
```

```python
from sklearn.cluster import KMeans
```

```python
plt.scatter(dt['Annual Income (k$)'],dt['Spending Score (1-100)'])
plt.xlabel('annual')
plt.ylabel('spending')
```

```python
km=KMeans(n_clusters=3)
```

```python
pr=km.fit_predict(x)
pr
```

```python
dt['cluster']=pr
```

```python
dt
```

```python
km.cluster_centers_
```

```python
df1=dt[dt['cluster']==0]
df2=dt[dt['cluster']==1]
df3=dt[dt['cluster']==2]
```

```python
plt.style.use('fast')
```

```python
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='black')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='blue')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='red')
```
